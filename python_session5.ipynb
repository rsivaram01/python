```python
# Inheritance:-
===============

creating a new class from a existing class is called Inheritance


super class , sub class

parent class , child class
```

```python
class parentclass:
    var1 = 'im var1'
    var2 = 'im var2'

class childclass(parentclass):
    var3 = 'im var3'
```

```python
pobj = parentclass()
```

```python
pobj.var1
```

```python
pobj.var2
```

```python
pobj.var3
```

```python
cobj = childclass()
```

```python
cobj.var3
```

```python
cobj.var1
```

```python
cobj.var2
```

```python
class dog:
    def __init__(self,name,age):
        self.aa = name
        self.bb = age
    def into(self):
        print(f"my dog name is {self.aa}")
        print(f"my dog age is {self.bb}")
        
    def sit(self):
        print(f"{self.aa} is sitting now")
    def turnarond(self):
        print(f"{self.aa} is turning around")
```

```python
dobj1 = dog('bruno',2)
```

```python
dobj1.into()
```

```python
dobj1.sit()
```

```python
dobj1.turnarond()
```
