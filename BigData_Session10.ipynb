```python
1. application:-
===============

Application is nothing but spark application, which is used to process data

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, count

# ----------------------------------------------------
# 1️⃣ Initialize Spark Session
# ----------------------------------------------------
spark = SparkSession.builder \
    .appName("EmployeeDataProcessing") \
    .master("local[*]") \
    .getOrCreate()

# ----------------------------------------------------
# 2️⃣ Read Input Data (CSV)
# ----------------------------------------------------
input_path = "hdfs:///data/employees.csv"  # or local: "data/employees.csv"
df = spark.read.option("header", True).option("inferSchema", True).csv(input_path)

print("=== Input Data ===")
df.show(5)  # this is an action 

# ----------------------------------------------------
# 3️⃣ Data Cleaning & Transformation
# ----------------------------------------------------
# Remove rows with null salary
df_clean = df.filter(col("salary").isNotNull())

# Calculate average salary by department
avg_salary_df = df_clean.groupBy("department").agg(
    avg("salary").alias("avg_salary"),
    count("*").alias("employee_count")
)

print("=== Average Salary by Department ===")
avg_salary_df.show()  # show is action
output_path = "hdfs:///output/employee_avg_salary"
avg_salary_df.write.mode("overwrite").parquet(output_path)

print(f"✅ Output written to {output_path}")


2. Job :-
========
When a job will create

whenever there is an action, that time a job will get create 
100 actions ----> 100 jobs

3. Stage:-
========
when a stage will be created 

whenever these a wide trasformation, there is stage
if your code 20 wide transformation -----> 20 stages


4. Task :-
=========
its equal to number stages * num of partitions 

a dataset ---> partitioned in 3 parts 
3 stages ----> under every stage it will create 3 tasks ---> 3*3 = 9 tasks


5. Shuffle :-
============
Shuffle is nothing but redistributing data across partitions

dataframe:-
dept  fine
==========
civil  100
cse    200
civil  200
ece    50

civil  100     ----> part1  stored in node 1


cse    200
civil  200    ------> part2  stored in node2


ece    50     -------> part3   stored in node3


sum(fine) group by dept 



civil 300
cse   200
ece    50





Transformation:-
===============
2 kind of transformation

1. Narrow transformation:-
-------------------------
narrow transformation are the one which dont's cause data suffling

i. filter
ii. withColumn
iii. casestament
iv. map()




2. wide transformation:-
----------------------
whenevr we apply transformation data, if that transfotmation cause data shufflign those kind of transformations called wide transformation

i. group by()
ii. distinct()
iii. repartition()




```
